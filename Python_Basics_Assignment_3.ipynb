# Assignment Questions

## Question 1: Enumerate with List

```python
fruits = ["apple", "banana", "cherry", "mango"]

for index, fruit in enumerate(fruits):
    print(f"{index} {fruit}")
```

## Question 2: Zip with Two Lists

```python
students = ["Rahim", "Karim", "Ayesha"]
scores = [85, 90, 78]

for student, score in zip(students, scores):
    print(f"{student} scored {score}")
```

## Question 3: Tuple and Loop

```python
subjects = ("Mathematics", "English", "Science", "History", "Art")

for subject in subjects:
    print(subject)
```

## Question 4: Set Operations

```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7}

common_elements = set1.intersection(set2)
unique_elements_set1 = set1.difference(set2)

print(f"Common elements: {common_elements}")
print(f"Unique elements of set1: {unique_elements_set1}")
```

## Question 5: Dictionary with Enumerate

```python
marks = {
    "Math": 80,
    "Physics": 75,
    "Chemistry": 85
}

for index, (subject, score) in enumerate(marks.items()):
    print(f"{index + 1} {subject}: {score}")
```
